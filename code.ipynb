```python
# -*- coding: utf-8 -*-
"""
ÂÜ∑ÈìæÈ£éÈô©‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÈ¢ÑÊµãÁ≥ªÁªü - Ê†∏ÂøÉÁâà
ÂàõÊñ∞ÁÇπÔºö22Áª¥È´òÁ∫ßÊó∂Â∫èÁâπÂæÅ + 24‰∏™‰∏âÁª¥‰∫§ÂèâÁâπÂæÅ + XGBoostÈõÜÊàêÂ≠¶‰π†
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from typing import Dict, List, Tuple, Optional
from dataclasses import dataclass, field
import logging
import warnings
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn. metrics import (classification_report, roc_auc_score, confusion_matrix,
                           roc_curve, f1_score, accuracy_score)
from sklearn.feature_selection import SelectKBest, f_classif
import xgboost as xgb
from xgboost import DMatrix
import seaborn as sns

try:
    from imblearn.combine import SMOTETomek
    IMBLEARN_AVAILABLE = True
except ImportError:
    IMBLEARN_AVAILABLE = False

# ÈÖçÁΩÆ
matplotlib.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
matplotlib. rcParams['axes.unicode_minus'] = False
warnings.filterwarnings('ignore')

# ========== ÈÖçÁΩÆÁÆ°ÁêÜ ==========
@dataclass
class SystemConfig:
    """Á≥ªÁªüÈÖçÁΩÆÁ±ª"""
    timesteps: int = 24
    window_stride: int = 6
    test_size: float = 0.2
    feature_selection_k: int = 30
    output_dir: str = "system_outputs"
    model_dir: str = "system_models"
    
    def __post_init__(self):
        for directory in [self.output_dir, self.model_dir]:
            Path(directory).mkdir(exist_ok=True)

# ========== Êï∞ÊçÆÂ§ÑÁêÜÂô® ==========
class DataProcessor: 
    """Êï∞ÊçÆÂ§ÑÁêÜÂô®"""
    
    def __init__(self, config: SystemConfig, logger: logging.Logger):
        self.config = config
        self.logger = logger
        self. scaler = RobustScaler()
        
    def load_and_preprocess(self, filepath: str) -> pd.DataFrame:
        """Âä†ËΩΩÂíåÈ¢ÑÂ§ÑÁêÜÊï∞ÊçÆ"""
        self.logger.info(f"üìÅ Âä†ËΩΩÊï∞ÊçÆ:  {filepath}")
        df = pd.read_csv(filepath)
        self.logger.info(f"   ÂéüÂßãÊï∞ÊçÆÂΩ¢Áä∂: {df. shape}")
        
        df = self._basic_cleaning(df)
        df = self._handle_outliers(df)
        df = self._feature_engineering(df)
        df = self._encode_labels(df)
        
        self.logger.info(f"   È¢ÑÂ§ÑÁêÜÂêéÂΩ¢Áä∂: {df.shape}")
        return df. sort_values('timestamp')
    
    def _basic_cleaning(self, df:  pd.DataFrame) -> pd.DataFrame:
        if 'timestamp' in df.columns:
            df['timestamp'] = pd.to_datetime(df['timestamp'])
        
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        for col in numeric_cols: 
            if df[col].isnull().sum() > 0:
                df[col] = df[col].fillna(df[col].median())
        return df
    
    def _handle_outliers(self, df:  pd.DataFrame) -> pd.DataFrame:
        if 'iot_temperature' in df.columns:
            q1, q99 = df['iot_temperature'].quantile([0.01, 0.99])
            df['iot_temperature'] = df['iot_temperature'].clip(q1, q99)
        
        constraints = {
            'route_risk_level': (0, 10),
            'supplier_reliability_score': (0, 100),
            'vehicle_gps_longitude': (-180, 180),
            'vehicle_gps_latitude': (-90, 90)
        }
        
        for col, (min_val, max_val) in constraints.items():
            if col in df.columns:
                df[col] = df[col].clip(min_val, max_val)
        return df
    
    def _feature_engineering(self, df: pd.DataFrame) -> pd.DataFrame:
        if 'timestamp' in df.columns:
            df['hour'] = df['timestamp'].dt. hour
            df['day_of_week'] = df['timestamp']. dt.dayofweek
            df['month'] = df['timestamp'].dt.month
            df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
            df['is_business_hours'] = ((df['hour'] >= 8) & (df['hour'] <= 18)).astype(int)
            df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
            df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
            df['dow_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
            df['dow_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
        return df
    
    def _encode_labels(self, df: pd.DataFrame) -> pd.DataFrame:
        if 'risk_classification' in df.columns:
            df['risk_label'] = (df['risk_classification'] == 'High Risk').astype(int)
            self.logger.info(f"   È£éÈô©ÂàÜÂ∏É: È´òÈ£éÈô©{df['risk_label'].sum()}‰∏™ ({df['risk_label'].mean():.1%})")
        return df

# ========== ÁâπÂæÅÊèêÂèñÂô®Ôºà22Áª¥È´òÁ∫ßÊó∂Â∫èÔºâ ==========
class AdvancedFeatureExtractor:
    """È´òÁ∫ßÁâπÂæÅÊèêÂèñÂô® - Ê†∏ÂøÉÂàõÊñ∞ÁÇπ1"""
    
    def __init__(self, config: SystemConfig, logger: logging.Logger):
        self.config = config
        self. logger = logger
        
    def extract_features(self, df: pd.DataFrame) -> Tuple[np.ndarray, ... ]:
        """ÊèêÂèñÁâπÂæÅ"""
        self.logger. info("üîç ÂºÄÂßãÁâπÂæÅÊèêÂèñ...")
        
        features = {
            'sequences': [],
            'static_features': [],
            'labels': [],
            'temperatures': []
        }
        
        n_samples = len(df)
        window_size = self.config.timesteps
        stride = self.config.window_stride
        
        for start_idx in range(0, n_samples - window_size + 1, stride):
            end_idx = start_idx + window_size
            window = df.iloc[start_idx:end_idx]
            
            if 'risk_label' not in window.columns or window['risk_label'].isna().any():
                continue
                
            seq_feat = self._extract_advanced_sequence_features(window)
            static_feat = self._extract_static_features(window)
            
            features['sequences'].append(seq_feat)
            features['static_features'].append(static_feat)
            features['labels'].append(window. iloc[0]['risk_label'])
            features['temperatures'].append(window['iot_temperature'].mean())
        
        result = [np.array(features[key]) for key in 
                 ['sequences', 'static_features', 'labels', 'temperatures']]
        
        self. logger.info(f"   ‚úÖ ÊèêÂèñÂÆåÊàê:  {len(result[0])}‰∏™Á™óÂè£")
        self.logger.info(f"   Â∫èÂàóÁâπÂæÅ: {result[0].shape[1]}Áª¥ (È´òÁ∫ßÊó∂Â∫èÁâπÂæÅ)")
        self.logger. info(f"   ÈùôÊÄÅÁâπÂæÅ: {result[1].shape[1]}Áª¥")
        
        return tuple(result)
    
    def _extract_advanced_sequence_features(self, window:  pd.DataFrame) -> np.ndarray:
        """ÊèêÂèñ22Áª¥È´òÁ∫ßÊó∂Â∫èÁâπÂæÅÔºàÂàõÊñ∞ÁÇπÔºâ"""
        temp_series = window['iot_temperature']. values
        n = len(temp_series)
        
        # Âü∫Á°ÄÁªüËÆ°Ôºà5Áª¥Ôºâ
        mean = temp_series.mean()
        std = temp_series.std() + 1e-8
        min_temp = temp_series.min()
        max_temp = temp_series.max()
        median = np.median(temp_series)
        
        # ÂàÜÂ∏ÉÂΩ¢ÊÄÅÔºà2Áª¥Ôºâ- ÂàõÊñ∞ÁâπÂæÅ
        skewness = pd.Series(temp_series).skew() if n > 2 else 0
        kurtosis = pd.Series(temp_series).kurtosis() if n > 3 else 0
        
        # ÊûÅÂ∑ÆÔºà1Áª¥Ôºâ
        temp_range = max_temp - min_temp
        
        # ÂÅèÁ¶ªÂ∫¶Ôºà2Áª¥Ôºâ- ÂàõÊñ∞ÁâπÂæÅ
        optimal_temp = 2.0
        deviation = np.mean(np.abs(temp_series - optimal_temp))
        max_deviation = np.max(np.abs(temp_series - optimal_temp))
        
        # ÂèòÂåñÁéáÔºà5Áª¥Ôºâ- ÂàõÊñ∞ÁâπÂæÅ
        changes = np.abs(np.diff(temp_series))
        avg_change = np.mean(changes) if len(changes) > 0 else 0
        max_change = np.max(changes) if len(changes) > 0 else 0
        min_change = np.min(changes) if len(changes) > 0 else 0
        change_90p = np.percentile(changes, 90) if len(changes) > 0 else 0
        change_10p = np.percentile(changes, 10) if len(changes) > 0 else 0
        
        # Âä†ÈÄüÂ∫¶Ôºà2Áª¥Ôºâ- ÂàõÊñ∞ÁâπÂæÅ
        if len(changes) > 1:
            accel = np.abs(np.diff(changes))
            max_accel = np.max(accel)
            avg_accel = np.mean(accel)
        else:
            max_accel = 0
            avg_accel = 0
        
        # Á®≥ÂÆöÊÄßÔºà1Áª¥Ôºâ
        stability = std / (abs(mean) + 1e-8)
        
        # Ë∂ãÂäøÔºà1Áª¥Ôºâ
        x = np.arange(n)
        trend = np.polyfit(x, temp_series, 1)[0] if n > 1 else 0
        
        # ÂõõÂàÜ‰ΩçÊï∞Ôºà2Áª¥Ôºâ
        q25 = np.percentile(temp_series, 25)
        q75 = np.percentile(temp_series, 75)
        
        return np.array([
            mean, std, min_temp, max_temp, median,           # 0-4
            skewness, kurtosis,                              # 5-6
            temp_range,                                      # 7
            deviation, max_deviation,                        # 8-9
            avg_change, max_change, min_change,              # 10-12
            change_90p, change_10p,                          # 13-14
            max_accel, avg_accel,                            # 15-16
            stability, trend,                                # 17-18
            q25, q75                                         # 19-20
        ])
    
    def _extract_static_features(self, window: pd.DataFrame) -> List[float]:
        """ÊèêÂèñÈùôÊÄÅÁâπÂæÅ"""
        first_row = window.iloc[0]
        temp_series = window['iot_temperature'].values
        
        supplier_score = first_row. get('supplier_reliability_score', 50)
        route_risk = first_row.get('route_risk_level', 5)
        hist_demand = first_row.get('historical_demand', 100)
        lead_time = first_row.get('lead_time_days', 3)
        
        min_temp = temp_series.min()
        max_temp = temp_series.max()
        avg_temp = temp_series.mean()
        temp_std = temp_series.std()
        
        cargo_type = self._infer_cargo_type(avg_temp)
        violation_metrics = self._calculate_violations(temp_series, cargo_type)
        risk_coeff = self._calculate_risk_coefficient(
            violation_metrics, supplier_score, route_risk
        )
        
        time_features = [
            first_row.get('hour_sin', 0),
            first_row.get('hour_cos', 0),
            first_row.get('dow_sin', 0),
            first_row.get('dow_cos', 0),
            first_row.get('is_weekend', 0),
            first_row.get('is_business_hours', 0)
        ]
        
        interactions = [
            supplier_score * route_risk,
            supplier_score * violation_metrics[0],
            route_risk * violation_metrics[0],
            np.log1p(hist_demand),
            np.sqrt(violation_metrics[0] + 1),
        ]
        
        return [
            supplier_score, lead_time, hist_demand, route_risk,
            cargo_type, min_temp, max_temp, avg_temp, temp_std,
            *violation_metrics, *time_features, *interactions, risk_coeff
        ]
    
    def _infer_cargo_type(self, avg_temp: float) -> int:
        if avg_temp < -10:
            return 2
        elif avg_temp < 5:
            return 0
        else:
            return 1
    
    def _calculate_violations(self, temps: np.ndarray, cargo_type: int) -> List[float]:
        ranges = {0: (2, 8), 1: (-1, 6), 2: (-20, -5)}
        min_temp, max_temp = ranges. get(cargo_type, (-1, 6))
        
        violations = (temps < min_temp) | (temps > max_temp)
        violation_count = violations.sum()
        violation_ratio = violation_count / len(temps)
        
        max_violation = 0
        if violation_count > 0:
            temp_violations = np.maximum(
                np.maximum(min_temp - temps, 0),
                np.maximum(temps - max_temp, 0)
            )
            max_violation = temp_violations.max()
        
        return [violation_count, violation_ratio, max_violation]
    
    def _calculate_risk_coefficient(self, violations: List[float], 
                                  supplier_score: float, route_risk: float) -> float:
        violation_count, violation_ratio, max_violation = violations
        time_dim = violation_ratio
        entity_dim = 1 - (supplier_score / 100)
        rule_dim = route_risk / 10
        base_risk = np.sqrt(time_dim) * entity_dim * (1 + rule_dim)
        violation_bonus = min(1.0, 0.1 * violation_count + 0.05 * max_violation)
        return base_risk + violation_bonus

# ========== ‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÂ∑•Á®ãÂô®ÔºàÊ†∏ÂøÉÂàõÊñ∞ÁÇπ2Ôºâ ==========
class ThreeDimensionalCrossFeatureEngineer:
    """‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÂ∑•Á®ãÂô® - Ê†∏ÂøÉÂàõÊñ∞ÁÇπ2"""
    
    def __init__(self, config: SystemConfig, logger: logging.Logger):
        self.config = config
        self.logger = logger
        self.feature_selector = None
        self.feature_names = []
        
    def create_cross_features(self, seq_features: np.ndarray, 
                            static_features: np.ndarray, 
                            labels: Optional[np.ndarray] = None) -> np.ndarray:
        """ÂàõÂª∫‰∏âÁª¥‰∫§ÂèâÁâπÂæÅ"""
        self.logger.info("üî® ‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÂ∑•Á®ã...")
        
        # ========== 1. Âü∫Á°ÄÁâπÂæÅÔºà10Áª¥Ôºâ ==========
        time_5_indices = [0, 1, 2, 3, 4]
        time_5 = seq_features[: , time_5_indices]
        time_5_names = ['TempMean', 'TempStd', 'TempMin', 'TempMax', 'TempMedian']
        
        entity_5_indices = [0, 3, 7, 12, 20]
        entity_5 = static_features[:, entity_5_indices]
        entity_5_names = ['Supplier', 'RouteRisk', 'AvgTemp', 'ViolationRatio', 'RiskCoeff']
        
        all_features = [time_5, entity_5]
        all_names = time_5_names + entity_5_names
        
        # ========== 2. ‰∫åÈò∂‰∫§ÂèâÔºàÊó∂Èó¥√óÂÆû‰ΩìÔºå25Áª¥Ôºâ ==========
        self.logger.info("   üìê ÁîüÊàê‰∫åÈò∂‰∫§ÂèâÁâπÂæÅ...")
        for i, t_name in enumerate(time_5_names):
            for j, e_name in enumerate(entity_5_names):
                cross_val = time_5[: , i] * entity_5[: , j]
                all_features.append(cross_val. reshape(-1, 1))
                all_names.append(f"{t_name}√ó{e_name}")
        
        # ========== 3. ‰∏âÈò∂‰∫§ÂèâÔºàÊó∂Èó¥√óÂÆû‰Ωì√óËßÑÂàôÔºå24Áª¥Ôºâ‚≠êÊ†∏ÂøÉÂàõÊñ∞ ==========
        self.logger. info("   ‚≠ê ÁîüÊàê‰∏âÈò∂‰∫§ÂèâÁâπÂæÅÔºàÊ†∏ÂøÉÂàõÊñ∞Ôºâ...")
        
        # Êó∂Èó¥Áª¥Â∫¶Ôºö4‰∏™È´òÁ∫ßÊó∂Â∫èÁâπÂæÅ
        time_advanced = {
            'TempStd': 1,
            'Skewness': 5,
            'TempDeviation': 8,
            'Kurtosis': 6
        }
        
        # ÂÆû‰ΩìÁª¥Â∫¶Ôºö2‰∏™‰æõÂ∫îÈìæÁâπÂæÅ
        entity_dim = {
            'Supplier': 0,
            'RouteRisk': 3
        }
        
        # ËßÑÂàôÁª¥Â∫¶Ôºö3‰∏™ËøùËßÑÁâπÂæÅ
        rule_dim = {
            'ViolationRatio': 12,
            'ViolationCount': 11,
            'MaxViolation': 13
        }
        
        cross_3rd_count = 0
        for t_name, t_idx in time_advanced.items():
            t_feat = seq_features[:, t_idx]
            
            for e_name, e_idx in entity_dim.items():
                e_feat = static_features[:, e_idx]
                
                for r_name, r_idx in rule_dim. items():
                    r_feat = static_features[:, r_idx]
                    
                    # ‰∏âÁª¥‰∫§ÂèâÔºöÊó∂Èó¥ √ó ÂÆû‰Ωì √ó ËßÑÂàô
                    cross_val = t_feat * e_feat * r_feat
                    all_features.append(cross_val.reshape(-1, 1))
                    all_names.append(f"{t_name}√ó{e_name}√ó{r_name}")
                    cross_3rd_count += 1
        
        self.logger.info(f"   ‚úÖ ‰∏âÈò∂‰∫§ÂèâÁâπÂæÅ:  {cross_3rd_count}‰∏™")
        self.logger.info(f"      (4Êó∂Èó¥ √ó 2ÂÆû‰Ωì √ó 3ËßÑÂàô = 24‰∏™)")
        
        # ========== 4. ÂêàÂπ∂ÁâπÂæÅ ==========
        all_features = np.concatenate(all_features, axis=1)
        self.logger.info(f"   ÂÄôÈÄâÁâπÂæÅÊÄªÊï∞: {all_features.shape[1]}Áª¥")
        self.logger.info(f"      (Âü∫Á°Ä10 + ‰∫åÈò∂25 + ‰∏âÈò∂{cross_3rd_count})")
        
        # ========== 5. ÁâπÂæÅÈÄâÊã© ==========
        if labels is not None and self.feature_selector is None:
            k = min(self.config.feature_selection_k, all_features.shape[1])
            self.feature_selector = SelectKBest(f_classif, k=k)
            all_features = self.feature_selector.fit_transform(all_features, labels)
            
            selected_mask = self.feature_selector.get_support()
            self.feature_names = [name for name, selected in zip(all_names, selected_mask) if selected]
            
            third_order_count = sum(1 for name in self.feature_names if name.count('√ó') == 2)
            self.logger.info(f"   ÁâπÂæÅÈÄâÊã©: Top {k}ÁâπÂæÅ")
            self.logger.info(f"   ÂÖ∂‰∏≠‰∏âÈò∂‰∫§Âèâ:  {third_order_count}‰∏™ËøõÂÖ•Top {k}")
            
        elif self.feature_selector is not None:
            all_features = self.feature_selector.transform(all_features)
        
        self.logger.info(f"‚úÖ ÊúÄÁªàÁâπÂæÅ:  {all_features.shape[1]}Áª¥")
        return all_features

# ========== XGBoostËÆ≠ÁªÉÂô® ==========
class XGBoostTrainer:
    """XGBoostËÆ≠ÁªÉÂô®"""
    
    def __init__(self, config: SystemConfig, logger:  logging.Logger):
        self.config = config
        self.logger = logger
        self.model = None
        self.scaler = StandardScaler()
        
    def train(self, X: np. ndarray, y: np.ndarray) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
        """ËÆ≠ÁªÉÊ®°Âûã"""
        self.logger. info("üöÄ XGBoostÊ®°ÂûãËÆ≠ÁªÉ...")
        
        # Ê†∑Êú¨Âπ≥Ë°°
        if IMBLEARN_AVAILABLE: 
            try:
                smotetomek = SMOTETomek(random_state=42)
                X_res, y_res = smotetomek.fit_resample(X, y)
                self.logger.info(f"   SMOTETomek:  {len(X)} ‚Üí {len(X_res)}Ê†∑Êú¨")
            except: 
                X_res, y_res = X, y
        else:
            X_res, y_res = X, y
        
        # ÂàÜÂâ≤Êï∞ÊçÆ
        X_train, X_val, y_train, y_val = train_test_split(
            X_res, y_res, test_size=self.config.test_size, 
            random_state=42, stratify=y_res
        )
        
        # Ê†áÂáÜÂåñ
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_val_scaled = self.scaler.transform(X_val)
        
        # Ê†∑Êú¨ÊùÉÈáç
        pos_ratio = y_train.sum() / len(y_train)
        scale_pos_weight = (1 - pos_ratio) / pos_ratio if pos_ratio > 0 else 1
        
        # ËÆ≠ÁªÉÂèÇÊï∞
        params = {
            'max_depth': 6,
            'eta': 0.1,
            'objective': 'binary:logistic',
            'eval_metric': ['logloss', 'auc'],
            'subsample': 0.8,
            'colsample_bytree': 0.8,
            'gamma': 1.0,
            'reg_alpha': 1.0,
            'reg_lambda': 5.0,
            'scale_pos_weight': scale_pos_weight,
            'random_state': 42
        }
        
        dtrain = DMatrix(X_train_scaled, label=y_train)
        dval = DMatrix(X_val_scaled, label=y_val)
        
        self.logger. info("   ÂºÄÂßãËÆ≠ÁªÉ...")
        self.model = xgb.train(
            params, dtrain,
            num_boost_round=200,
            evals=[(dtrain, 'train'), (dval, 'val')],
            early_stopping_rounds=20,
            verbose_eval=20
        )
        
        y_proba = self.model.predict(dval)
        y_pred = (y_proba > 0.5).astype(int)
        
        self._log_metrics(y_val, y_pred, y_proba)
        
        return X_val_scaled, y_val, y_proba
    
    def _log_metrics(self, y_true, y_pred, y_proba):
        self.logger.info("\n" + "="*60)
        self.logger.info("üìä Ê®°ÂûãÊÄßËÉΩËØÑ‰º∞")
        self.logger.info("="*60)
        
        acc = accuracy_score(y_true, y_pred)
        auc = roc_auc_score(y_true, y_proba)
        f1 = f1_score(y_true, y_pred)
        
        self.logger. info(f"ÂáÜÁ°ÆÁéá (Accuracy): {acc:.4f}")
        self.logger.info(f"AUCÂæóÂàÜ:  {auc:.4f}")
        self.logger.info(f"F1ÂæóÂàÜ: {f1:.4f}")
        self.logger.info(f"\n{classification_report(y_true, y_pred, target_names=['‰ΩéÈ£éÈô©', 'È´òÈ£éÈô©'])}")
    
    def get_feature_importance(self, feature_names: List[str]) -> pd.DataFrame:
        """Ëé∑ÂèñÁâπÂæÅÈáçË¶ÅÊÄß"""
        importance = self.model.get_score(importance_type='weight')
        
        importance_list = []
        for i, name in enumerate(feature_names):
            score = importance. get(f"f{i}", 0)
            importance_list.append({'feature': name, 'importance': score})
        
        df = pd.DataFrame(importance_list)
        df = df.sort_values('importance', ascending=False).reset_index(drop=True)
        return df

# ========== ÂèØËßÜÂåñÂô® ==========
class Visualizer:
    """ÂèØËßÜÂåñÂô®"""
    
    def __init__(self, config: SystemConfig, logger: logging.Logger):
        self.config = config
        self.logger = logger
        
    def plot_feature_importance(self, importance_df: pd.DataFrame, top_n: int = 30):
        """ÁâπÂæÅÈáçË¶ÅÊÄßÂèØËßÜÂåñÔºàÁ™ÅÂá∫‰∏âÈò∂‰∫§ÂèâÔºâ"""
        top_features = importance_df.head(top_n).copy()
        
        # ÁâπÂæÅÂàÜÁ±ª
        def classify_feature(name):
            if '√ó' in name:
                if name.count('√ó') == 2:
                    return '‰∏âÈò∂‰∫§Âèâ'
                elif name.count('√ó') == 1:
                    return '‰∫åÈò∂‰∫§Âèâ'
            
            time_keywords = ['Temp', 'Skew', 'Kurt', 'Change', 'Accel', 
                           'Stability', 'Trend', 'Deviation', 'Range', 'Q']
            if any(kw in name for kw in time_keywords):
                return 'Êó∂Èó¥Áª¥Â∫¶'
            
            return 'ÂÆû‰Ωì/ËßÑÂàô'
        
        top_features['type'] = top_features['feature']. apply(classify_feature)
        
        # ÁªüËÆ°
        type_counts = top_features['type'].value_counts()
        third_order = type_counts. get('‰∏âÈò∂‰∫§Âèâ', 0)
        second_order = type_counts.get('‰∫åÈò∂‰∫§Âèâ', 0)
        time_dim = type_counts.get('Êó∂Èó¥Áª¥Â∫¶', 0)
        entity_dim = type_counts.get('ÂÆû‰Ωì/ËßÑÂàô', 0)
        
        self.logger.info(f"\n{'='*60}")
        self.logger.info(f"üìä ÁâπÂæÅÈáçË¶ÅÊÄßÂàÜÊûê (Top {top_n})")
        self.logger.info(f"{'='*60}")
        self.logger.info(f"‚≠ê ‰∏âÈò∂‰∫§ÂèâÁâπÂæÅ: {third_order}‰∏™ ({third_order/top_n:.1%})")
        self.logger.info(f"   ‰∫åÈò∂‰∫§ÂèâÁâπÂæÅ: {second_order}‰∏™ ({second_order/top_n:.1%})")
        self.logger.info(f"   Êó∂Èó¥Áª¥Â∫¶ÁâπÂæÅ: {time_dim}‰∏™ ({time_dim/top_n:.1%})")
        self.logger.info(f"   ÂÆû‰Ωì/ËßÑÂàôÁâπÂæÅ: {entity_dim}‰∏™ ({entity_dim/top_n:.1%})")
        
        # ÁªòÂõæ
        plt.figure(figsize=(14, 10))
        
        colors = []
        for ftype in top_features['type']: 
            if ftype == '‰∏âÈò∂‰∫§Âèâ':
                colors.append('#E74C3C')  # Á∫¢Ëâ≤
            elif ftype == '‰∫åÈò∂‰∫§Âèâ': 
                colors.append('#F39C12')  # Ê©ôËâ≤
            elif ftype == 'Êó∂Èó¥Áª¥Â∫¶':
                colors.append('#3498DB')  # ËìùËâ≤
            else:
                colors.append('#95A5A6')  # ÁÅ∞Ëâ≤
        
        y_pos = np.arange(len(top_features))
        plt.barh(y_pos, top_features['importance'], color=colors, alpha=0.8, 
                edgecolor='black', linewidth=0.5)
        
        labels = []
        for idx, row in top_features.iterrows():
            marker = "‚≠ê" if row['type'] == '‰∏âÈò∂‰∫§Âèâ' else "  "
            labels.append(f"{marker}{row['feature']}")
        
        plt.yticks(y_pos, labels, fontsize=9)
        plt.xlabel('ÈáçË¶ÅÊÄßÂæóÂàÜ (XGBoost Weight)', fontsize=12)
        plt.title(f'ÁâπÂæÅÈáçË¶ÅÊÄß Top{top_n}\n' + 
                 f'Á∫¢Ëâ≤=‰∏âÈò∂‰∫§Âèâ({third_order}) | Ê©ôËâ≤=‰∫åÈò∂‰∫§Âèâ({second_order}) | ' +
                 f'ËìùËâ≤=Êó∂Èó¥Áª¥Â∫¶({time_dim}) | ÁÅ∞Ëâ≤=ÂÆû‰Ωì/ËßÑÂàô({entity_dim})',
                 fontsize=13, fontweight='bold')
        plt.gca().invert_yaxis()
        plt.grid(True, axis='x', alpha=0.3)
        
        from matplotlib.patches import Patch
        legend_elements = [
            Patch(facecolor='#E74C3C', label=f'‚≠ê ‰∏âÈò∂‰∫§Âèâ ({third_order}‰∏™, {third_order/top_n:.0%})'),
            Patch(facecolor='#F39C12', label=f'‰∫åÈò∂‰∫§Âèâ ({second_order}‰∏™, {second_order/top_n:.0%})'),
            Patch(facecolor='#3498DB', label=f'Êó∂Èó¥Áª¥Â∫¶ ({time_dim}‰∏™, {time_dim/top_n:.0%})'),
            Patch(facecolor='#95A5A6', label=f'ÂÆû‰Ωì/ËßÑÂàô ({entity_dim}‰∏™, {entity_dim/top_n:.0%})')
        ]
        plt. legend(handles=legend_elements, loc='lower right', fontsize=10)
        
        plt.tight_layout()
        plt.savefig(f"{self.config.output_dir}/feature_importance.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
        
        # ËØ¶ÁªÜÂàóË°®
        self.logger.info(f"\n{'='*60}")
        self.logger.info(f"üìã ÁâπÂæÅÈáçË¶ÅÊÄßËØ¶ÁªÜÊéíÂêç")
        self.logger.info(f"{'='*60}")
        for idx, row in top_features.iterrows():
            marker = "‚≠ê‚≠ê‚≠ê" if row['type'] == '‰∏âÈò∂‰∫§Âèâ' else "    "
            type_tag = f"[{row['type']}]".ljust(10)
            self.logger.info(f"{idx+1:2d}. {marker} {type_tag} {row['feature']:40s} {row['importance']:6.1f}")
        
        # ‰∏âÈò∂‰∫§ÂèâÁâπÂæÅÊ±áÊÄª
        if third_order > 0:
            self.logger.info(f"\n{'='*60}")
            self.logger.info(f"üéØ ‰∏âÈò∂‰∫§ÂèâÁâπÂæÅÊ±áÊÄª ({third_order}‰∏™ËøõÂÖ•Top {top_n})")
            self.logger.info(f"{'='*60}")
            third_features = top_features[top_features['type'] == '‰∏âÈò∂‰∫§Âèâ']
            for idx, row in third_features.iterrows():
                rank = top_features.index.get_loc(idx) + 1
                self.logger.info(f"ÊéíÂêç{rank:2d}:{row['feature']:50s} (ÂæóÂàÜ={row['importance']:.1f})")
        
        return top_features
    
    def plot_confusion_matrix(self, y_true:  np.ndarray, y_pred: np.ndarray):
        """Ê∑∑Ê∑ÜÁü©Èòµ"""
        cm = confusion_matrix(y_true, y_pred)
        
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=['‰ΩéÈ£éÈô©', 'È´òÈ£éÈô©'],
                   yticklabels=['‰ΩéÈ£éÈô©', 'È´òÈ£éÈô©'])
        plt.title('Ê∑∑Ê∑ÜÁü©Èòµ', fontsize=14, fontweight='bold')
        plt.xlabel('È¢ÑÊµãÊ†áÁ≠æ')
        plt.ylabel('ÁúüÂÆûÊ†áÁ≠æ')
        plt.tight_layout()
        plt.savefig(f"{self.config.output_dir}/confusion_matrix.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
    
    def plot_roc_curve(self, y_true: np.ndarray, y_proba: np.ndarray):
        """ROCÊõ≤Á∫ø"""
        fpr, tpr, _ = roc_curve(y_true, y_proba)
        auc = roc_auc_score(y_true, y_proba)
        
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, linewidth=2, label=f'ROCÊõ≤Á∫ø (AUC = {auc:.3f})', color='#3498DB')
        plt.plot([0, 1], [0, 1], 'k--', linewidth=1, label='ÈöèÊú∫ÂàÜÁ±ªÂô®')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('ÂÅáÊ≠£Áéá (False Positive Rate)')
        plt.ylabel('ÁúüÊ≠£Áéá (True Positive Rate)')
        plt.title('ROCÊõ≤Á∫ø', fontsize=14, fontweight='bold')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(f"{self.config.output_dir}/roc_curve.png", 
                   dpi=300, bbox_inches='tight')
        plt.show()
        
        return auc

# ========== ‰∏ªÂáΩÊï∞ ==========
def main():
    logging.basicConfig(
        level=logging.INFO,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )
    logger = logging.getLogger(__name__)
    
    logger.info("="*70)
    logger.info("üöÄ ÂÜ∑ÈìæËøêËæìÈ£éÈô©È¢ÑÊµãÁ≥ªÁªü")
    logger.info("   Ê†∏ÂøÉÂàõÊñ∞:  ‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÂ∑•Á®ã + È´òÁ∫ßÊó∂Â∫èÁâπÂæÅ + XGBoost")
    logger.info("="*70)
    
    config = SystemConfig()
    
    # Ê≠•È™§1: Êï∞ÊçÆÈ¢ÑÂ§ÑÁêÜ
    logger.info("\n" + "="*70)
    logger.info("„ÄêÊ≠•È™§1/5„ÄëÊï∞ÊçÆÈ¢ÑÂ§ÑÁêÜ")
    logger.info("="*70)
    processor = DataProcessor(config, logger)
    df = processor.load_and_preprocess('Êï∞ÊçÆÈõÜ.csv')
    
    # Ê≠•È™§2: ÁâπÂæÅÊèêÂèñ
    logger.info("\n" + "="*70)
    logger.info("„ÄêÊ≠•È™§2/5„ÄëÈ´òÁ∫ßÁâπÂæÅÊèêÂèñÔºà22Áª¥Êó∂Â∫èÁâπÂæÅÔºâ")
    logger.info("="*70)
    extractor = AdvancedFeatureExtractor(config, logger)
    seq_features, static_features, labels, temperatures = extractor.extract_features(df)
    
    logger.info(f"\nüìä Êï∞ÊçÆÈõÜÁªüËÆ°:")
    logger.info(f"   ÊÄªÊ†∑Êú¨:  {len(labels)}‰∏™Á™óÂè£")
    logger.info(f"   È´òÈ£éÈô©ÊØî‰æã: {labels.mean():.1%}")
    logger.info(f"   Ê∏©Â∫¶ËåÉÂõ¥: {temperatures.min():.1f}¬∞C ~ {temperatures.max():.1f}¬∞C")
    
    # Ê≠•È™§3: ‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÂ∑•Á®ã
    logger.info("\n" + "="*70)
    logger.info("„ÄêÊ≠•È™§3/5„Äë‰∏âÁª¥‰∫§ÂèâÁâπÂæÅÂ∑•Á®ãÔºàÊ†∏ÂøÉÂàõÊñ∞Ôºâ")
    logger.info("="*70)
    cross_engineer = ThreeDimensionalCrossFeatureEngineer(config, logger)
    X = cross_engineer.create_cross_features(seq_features, static_features, labels)
    
    # Ê≠•È™§4: Ê®°ÂûãËÆ≠ÁªÉ
    logger.info("\n" + "="*70)
    logger.info("„ÄêÊ≠•È™§4/5„ÄëXGBoostÊ®°ÂûãËÆ≠ÁªÉ")
    logger.info("="*70)
    trainer = XGBoostTrainer(config, logger)
    X_val, y_val, y_proba = trainer.train(X, labels)
    
    # Ê≠•È™§5: ÁªìÊûúÂàÜÊûê‰∏éÂèØËßÜÂåñ
    logger.info("\n" + "="*70)
    logger.info("„ÄêÊ≠•È™§5/5„ÄëÁªìÊûúÂàÜÊûê‰∏éÂèØËßÜÂåñ")
    logger.info("="*70)
    
    # ÁâπÂæÅÈáçË¶ÅÊÄßÂàÜÊûê
    importance_df = trainer.get_feature_importance(cross_engineer.feature_names)
    
    # ÂèØËßÜÂåñ
    visualizer = Visualizer(config, logger)
    visualizer.plot_feature_importance(importance_df, top_n=30)
    
    y_pred = (y_proba > 0.5).astype(int)
    visualizer.plot_confusion_matrix(y_val, y_pred)
    auc_score = visualizer.plot_roc_curve(y_val, y_proba)
    
    # ‰øùÂ≠òÁªìÊûú
    importance_df.to_csv(f"{config.output_dir}/feature_importance.csv", 
                        index=False, encoding='utf-8-sig')
    
    # ÊúÄÁªàÊÄªÁªì
    logger.info("\n" + "="*70)
    logger.info("‚úÖ Á≥ªÁªüËøêË°åÂÆåÊàê")
    logger.info("="*70)
    logger.info(f"üìä Ê®°ÂûãÊÄßËÉΩ:")
    logger.info(f"   AUC: {auc_score:.4f}")
    logger.info(f"   ÂáÜÁ°ÆÁéá: {accuracy_score(y_val, y_pred):.4f}")
    logger.info(f"   F1ÂæóÂàÜ: {f1_score(y_val, y_pred):.4f}")
    
    third_order_count = sum(1 for name in cross_engineer.feature_names if name.count('√ó') == 2)
    logger.info(f"\nüéØ ÂàõÊñ∞ÁÇπÈ™åËØÅ:")
    logger.info(f"   ‰∏âÈò∂‰∫§ÂèâÁâπÂæÅ: {third_order_count}‰∏™ËøõÂÖ•Top 30")
    logger.info(f"   Âç†ÊØî: {third_order_count/30:.1%}")
    logger.info(f"\nüíæ ÁªìÊûúÂ∑≤‰øùÂ≠òËá≥: {config.output_dir}/")
    logger.info("="*70)

if __name__ == "__main__":
    main()
```
